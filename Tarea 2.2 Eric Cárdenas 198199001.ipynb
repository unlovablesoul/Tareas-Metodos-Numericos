#### Escriba un código que tome un array producido con np.zeros(100) y produzca el mismo resultado que la función np.linspace(0,0.1,100), sin usar esa función.

```python
import numpy as np

valores = 100
array = np.zeros(valores)
#definimos un contador

i = 0
m =0.1 #iteracion de datos
f = valores - 1 #total de datos del array
num = 0 #valor inicial de los datos


print(np.linspace(0,0.1,valores)) #<-- para comprobar quitar el comentario

#usando while
while i < valores:
    array[i]=num
    num=num + m/f #m/f es la relación entre la itinerancia de datos sobre el valor total de datos
    i=i+1
print("*",array) # * es el valor del array usando otro codigo
```
